#### 1st. GridSearchCV
- max_depth, gamma

```python
my_scorer_f1 = make_scorer(f1_score, greater_is_better = True)
my_scorer_acc = make_scorer(accuracy_score, greater_is_better = True)
```

```python
xgb_1 = XGBClassifier(
 learning_rate =0.1,
 n_estimators=100,
)
```

```python
xgb_params_1 = {
    'max_depth' : [3,5,7],
    'gamma' : [0, 0.1],
} 
```
